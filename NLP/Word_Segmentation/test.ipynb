```python
import re
import csv

csvFile = open("test.csv", encoding='utf-8',)
reader = csv.reader(csvFile)
# 建立空字典
senlst = []
for item in reader:
    # 忽略第一行
    if reader.line_num == 1:
        continue
    senlst.append(item[1])
csvFile.close()

a_lst=[]
for lst in senlst:
    relst=re.findall('\d+年\d+月|\d+时\d+分|\d+[周秒块岁度颗种天处亿位号套元打类户项省日步世次口米分家月座台个倍部时吨国对年县篇局亩人段％条张名卷所页点万代辆匹‰]|\d+公顷|\d+万亩|\d+亿亩',lst)
    for i in range(len(relst)):
        a_lst.append(relst[i])
dic_lst=list(set(a_lst))
print(dic_lst)

csvFile = open("./OOV_dict.csv", "w",newline='')
writer = csv.writer(csvFile)
for item in dic_lst:
    writer.writerow([item])
csvFile.close()
```

```python
import re
import csv

csvFile = open("test.csv", encoding='utf-8',)
reader = csv.reader(csvFile)
# 建立空字典
senlst = []
for item in reader:
    # 忽略第一行
    if reader.line_num == 1:
        continue
    senlst.append(item[1])
csvFile.close()

a_lst=[]
for lst in senlst:
    relst=re.findall('\d+[^_]{2}',lst)
    for i in range(len(relst)):
        a_lst.append(relst[i][-2:])
dic_lst=list(set(a_lst))
print(dic_lst)
```
