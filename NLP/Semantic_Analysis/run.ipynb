```python
import csv 
import re
# 读取csv至字典
csvFile = open("sentence.csv", encoding='utf-8')
reader = csv.reader(csvFile)
# 建立空列表
senlst = []
for item in reader:
    if item:
        senlst.append(item[0])
csvFile.close()
```

```python
lst=senlst[6]
print(lst)
print((re.sub('\/[a-zA-Z]+\s|[a-zA-Z]+\s|\/[a-zA-Z]+','',lst).split(' ')[0:-1]))
print(re.findall('\/[a-zA-Z]+',lst))
print(len(re.sub('\/[a-zA-Z]+\s|[a-zA-Z]+\s|\/[a-zA-Z]+','',lst).split(' ')[0:-1]))
print(len(re.findall('\/[a-zA-Z]+',lst)))
```

```python
def tran(lst):
    lsta=re.sub('\/[a-zA-Z]+\s|[a-zA-Z]+\s|\/[a-zA-Z]+','',lst).split(' ')[0:-1]
    lstb=re.findall('\/[a-zA-Z]+',lst)
    return lsta,lstb

lst1=[]
lst2=[]
for item in senlst:
    lsta,lstb=tran(item)
    lst1.append(lsta)
    lst2.append(lstb)
print(lst1[100])
print(lst2[100])
print(len(lst1))
print(len(lst2))
```

```python
csvFile = open("lst1.csv", "w",newline='',encoding='utf-8')
writer = csv.writer(csvFile)
for item in lst1:
    writer.writerow(item)
csvFile.close()

csvFile = open("lst2.csv", "w",newline='')
writer = csv.writer(csvFile)
for item in lst2:
    writer.writerow(item)
csvFile.close()
```
