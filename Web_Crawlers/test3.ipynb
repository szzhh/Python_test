```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
kv = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}

url = 'https://www.shanghairanking.cn/rankings/bcur/2022'
r = requests.get(url, headers = kv)
r.encoding = r.apparent_encoding
r.status_code
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
import pandas as pd
```

```python
for sp in soup.find('tbody',attrs={'data-v-3fe7d390':""}).find_all('tr',attrs={'data-v-3fe7d390':""},recursive=False):
    rank = sp.find('div', class_='ranking').text.strip()
    name_cn = sp.find('a', class_='name-cn').text
    name_en = sp.find('a', class_='name-en').text
    tags = sp.find('p', class_='tags').text
    ad = sp.find_all('td',attrs={'data-v-3fe7d390':""})[2].text.strip()
    clas = sp.find_all('td',attrs={'data-v-3fe7d390':""})[3].text.strip()
    score = sp.find_all('td',attrs={'data-v-3fe7d390':""})[4].text.strip()
    print(rank,name_cn,name_en,tags,ad,clas,score)
    df = pd.DataFrame(data=[[rank,name_cn,name_en,tags,ad,clas,score]])
    if rank == '1':
        df.to_csv('./中国大学排名.csv', header=['排名', '中文名','英文名', '标签','省市', '类型', '总分'], index=False, mode='a')
    else:
        df.to_csv('./中国大学排名.csv', header=False, index=False, mode='a')
```
